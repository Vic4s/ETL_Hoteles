```python
import pandas as pd 
import numpy as np 
import requests
```

```python
endpoint = "https://datos.madrid.es/egob/catalogo/300107-0-agenda-actividades-eventos.json"
```

```python
# Realizar la solicitud GET
response = requests.get(endpoint)
```

```python
if response.status_code == 200:
    data = response.json()  # Convertir la respuesta a un diccionario de Python
else:
    print(f"Error en la solicitud: {response.status_code}")
```

```python
id_evento = int(data.get("@graph")[0].get("id")) if data.get("@graph")[0].get("id") != "" else np.nan
print(id_evento)

nombre_evento = data.get("@graph")[0].get("title") if data.get("@graph")[0].get("title") != "" else "Sin nombre"
print(nombre_evento)

url_evento = data.get("@graph")[0].get("link") if data.get("@graph")[0].get("link") != "" else "Sin link"
print(url_evento)

codigo_postal = data.get("@graph")[0].get("address").get("area").get("postal-code") if data.get("@graph")[0].get("address").get("area").get("postal-code") != "" else "Sin CP"
print(codigo_postal)

direccion = data.get("@graph")[0].get("address").get("area").get("street-address") if data.get("@graph")[0].get("address").get("area").get("street-address") != "" else "Sin direccion"
print(direccion)

horario = data.get("@graph")[0].get("time") if data.get("@graph")[0].get("time") != "" else pd.NaT
print(horario)

fecha_inicio = pd.to_datetime(data.get("@graph")[0].get("dtstart")).strftime('%Y-%m-%d') if data.get("@graph")[0].get("dtstart") != "" else pd.NaT
print(fecha_inicio)

fecha_fin = pd.to_datetime(data.get("@graph")[0].get("dtend")).strftime('%Y-%m-%d') if data.get("@graph")[0].get("dtstart") != "" else pd.NaT
print(fecha_fin)

organizacion = data.get("@graph")[0].get("organization").get("organization-name") if data.get("@graph")[0].get("organization").get("organization-name") != "" else "Sin organizacion"
print(organizacion)

id_ciudad = 1
```

# HACER LA MEDIA DE LAS ESTRELLAS!!!! <-- TODO !!!!!


```python
eventos = data.get("@graph")

fecha_estand = pd.to_datetime("2025-03-02")

dictio_eventos = {
        "id_evento": [],
        "nombre_evento": [],
        "url_evento": [],
        "codigo_postal": [],
        "direccion": [],
        "horario": [],
        "fecha_inicio": [],
        "fecha_fin": [],
        "organizacion": [],
        "id_ciudad": []
    }

for i, evento in enumerate(eventos):

    graph = data.get("@graph")


    if (pd.to_datetime(graph[i].get("dtstart")) <= fecha_estand) & (fecha_estand <= pd.to_datetime(graph[i].get("dtend"))):
        
        id_evento = int(graph[i].get("id", np.nan)) 
        dictio_eventos["id_evento"].append(id_evento)

        nombre_evento = graph[i].get("title", "Sin nombre") 
        dictio_eventos["nombre_evento"].append(nombre_evento)

        url_evento = graph[i].get("link", "Sin link") 
        dictio_eventos["url_evento"].append(url_evento)

        codigo_postal = int(graph[i].get("address", {}).get("area", {}).get("postal-code", 0))
        dictio_eventos["codigo_postal"].append(codigo_postal)

        direccion = graph[i].get("address", {}).get("area", {}).get("street-address", "Sin direccion") 
        dictio_eventos["direccion"].append(direccion)

        horario = graph[i].get("time", "Sin horario") if graph[i].get("time", "Sin horario") != "" else "Sin horario"
        dictio_eventos["horario"].append(horario)

        fecha_inicio = pd.to_datetime(graph[i].get("dtstart", pd.NaT)).strftime('%Y-%m-%d')
        dictio_eventos["fecha_inicio"].append(fecha_inicio)

        fecha_fin = pd.to_datetime(graph[i].get("dtend", pd.NaT)).strftime('%Y-%m-%d') 
        dictio_eventos["fecha_fin"].append(fecha_fin)

        organizacion = graph[i].get("organization", {}).get("organization-name", "Sin organizacion") 
        dictio_eventos["organizacion"].append(organizacion)

        dictio_eventos["id_ciudad"].append(1)


```

```python
df_eventos = pd.DataFrame(dictio_eventos)
df_eventos.sample(25)
```

```python
nulos_porcentaje = round((df_eventos.isnull().sum() / len(df_eventos)) * 100, 2)
nulos_porcentaje = nulos_porcentaje[nulos_porcentaje > 0]
print(nulos_porcentaje)
```
